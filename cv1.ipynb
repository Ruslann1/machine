```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimage
from PIL import Image


img = Image.open('6_26.jpg')
r, g, b = img.split()
im1 = Image.merge("RGB", (r, g, b))
im2 = Image.merge("RGB", (r, b, g))
im3 = Image.merge("RGB", (g, b, r))
im4 = Image.merge("RGB", (g, r, b))
im5 = Image.merge("RGB", (b, g, r))
im6 = Image.merge("RGB", (b, r, g))

plt.subplot(231);
plt.imshow(im1);
plt.subplot(232);
plt.imshow(im2);
plt.subplot(233);
plt.imshow(im3);
plt.subplot(234);
plt.imshow(im4);
plt.subplot(235);
plt.imshow(im5);
plt.subplot(236);
plt.imshow(im6);
plt.show();
```
